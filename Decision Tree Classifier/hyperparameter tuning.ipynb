# Hyperparameter tuning

Per trovare i migliori parametri di un albero di classificazione, uno degli approcci più comuni e efficaci è l'uso della ricerca iperparametrica (hyperparameter tuning). Questo processo consiste nel testare diverse combinazioni di parametri e valutare le prestazioni del modello su un insieme di validazione. Ecco i passaggi principali per trovare i parametri migliori di un albero di classificazione:

### 1. Selezione degli Iperparametri da Ottimizzare

Gli iperparametri comuni di un albero di classificazione (Decision Tree) includono:

- `criterion`: il criterio di divisione (ad esempio, "gini" o "entropy").
- `max_depth`: la profondità massima dell'albero.
- `min_samples_split`: il numero minimo di campioni richiesti per suddividere un nodo.
- `min_samples_leaf`: il numero minimo di campioni che deve contenere un nodo foglia.
- `max_features`: il numero massimo di caratteristiche da considerare per suddividere un nodo.

### 3. Scelta del Metodo di Ricerca

Ci sono vari metodi per effettuare la ricerca iperparametrica:

- **Grid Search**: Prova tutte le combinazioni possibili di iperparametri.
- **Random Search**: Prova un numero predefinito di combinazioni casuali di iperparametri.
- **Bayesian Optimization**: Usa metodi di ottimizzazione bayesiana per trovare la combinazione ottimale di iperparametri.

Una volta trovati i migliori parametri, valutali su un insieme di test per verificare le prestazioni del modello con i dati non visti.

### Considerazioni Finali

- **Validazione incrociata**: Usare la validazione incrociata (cross-validation) aiuta a ridurre il rischio di overfitting.
- **Bilanciamento del tempo e delle risorse**: La Grid Search può essere molto costosa in termini di tempo e risorse computazionali. Per dataset e modelli più complessi, considera l'uso di Random Search o ottimizzazione bayesiana.

Questo processo ti aiuterà a trovare i migliori iperparametri per il tuo albero di classificazione e migliorare le prestazioni del tuo modello.

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Caricamento del dataset e suddivisione in train e test
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

# Definizione dei parametri di ricerca
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Definizione del modello
clf = DecisionTreeClassifier()

# Definizione della ricerca con validazione incrociata
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, scoring='accuracy')

# Esecuzione della ricerca
grid_search.fit(X_train, y_train)

# Stampa dei migliori parametri
print("Migliori parametri trovati:", grid_search.best_params_)

# Valutazione del modello sul set di test
best_clf = grid_search.best_estimator_
accuracy = best_clf.score(X_test, y_test)
print("Accuratezza sul set di test:", accuracy)
````

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Caricamento del dataset
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

# Definizione dei parametri di ricerca
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Definizione del modello
clf = DecisionTreeClassifier()

# Definizione della ricerca con validazione incrociata
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, scoring='accuracy')

# Esecuzione della ricerca
grid_search.fit(X_train, y_train)

import pandas as pd

# Estrazione dei risultati
results = pd.DataFrame(grid_search.cv_results_)

# Visualizzazione delle prime righe del dataframe
print(results.head())

import matplotlib.pyplot as plt

# Filtriamo i risultati per visualizzare le prestazioni rispetto a max_depth
depth_results = results[results['param_max_depth'].notnull()]

# Creazione del grafico
plt.figure(figsize=(10, 6))
for criterion in param_grid['criterion']:
    subset = depth_results[depth_results['param_criterion'] == criterion]
    plt.plot(subset['param_max_depth'], subset['mean_test_score'], label=f'criterion: {criterion}')
    
plt.xlabel('Max Depth')
plt.ylabel('Mean Test Score')
plt.title('Decision Tree Performance by Max Depth and Criterion')
plt.legend()
plt.grid(True)
plt.show()

# Filtriamo i risultati per visualizzare le prestazioni rispetto a min_samples_split
split_results = results[results['param_min_samples_split'].notnull()]

# Creazione del grafico
plt.figure(figsize=(10, 6))
for criterion in param_grid['criterion']:
    subset = split_results[split_results['param_criterion'] == criterion]
    plt.plot(subset['param_min_samples_split'], subset['mean_test_score'], label=f'criterion: {criterion}')
    
plt.xlabel('Min Samples Split')
plt.ylabel('Mean Test Score')
plt.title('Decision Tree Performance by Min Samples Split and Criterion')
plt.legend()
plt.grid(True)
plt.show()

import seaborn as sns

# Creazione di una heatmap per visualizzare l'interazione tra max_depth e min_samples_split
heatmap_data = results.pivot_table(values='mean_test_score', index='param_max_depth', columns='param_min_samples_split')

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='viridis')
plt.title('Decision Tree Performance by Max Depth and Min Samples Split')
plt.xlabel('Min Samples Split')
plt.ylabel('Max Depth')
plt.show()
````

### 1. `ccp_alpha`: 0.0

- **Descrizione**: Questo parametro rappresenta il costo della complessità della potatura. Viene utilizzato per post-pruning dell'albero.
- **Funzione**: Se impostato a un valore maggiore di zero, il parametro controlla la quantità di potatura applicata all'albero. Maggiore è il valore, più rami verranno potati.
- **Valore di Default**: 0.0 (nessuna potatura).

### 2. `class_weight`: None

- **Descrizione**: Pesi associati alle classi. Utilizzato per gestire classi sbilanciate.
- **Funzione**: Se impostato a 'balanced', i pesi sono automaticamente regolati inversamente proporzionali alla frequenza delle classi nel dataset di input.
- **Valore di Default**: None (tutte le classi sono considerate con lo stesso peso).

### 3. `criterion`: 'gini'

- **Descrizione**: Il criterio utilizzato per misurare la qualità delle suddivisioni.
- **Funzione**: Determina come vengono selezionati i nodi di divisione. Può essere 'gini' (indice di Gini) o 'entropy' (entropia informativa).
- **Valore di Default**: 'gini'.

### 4. `max_depth`: None

- **Descrizione**: La profondità massima dell'albero.
- **Funzione**: Limita la crescita dell'albero a una certa profondità per prevenire l'overfitting.
- **Valore di Default**: None (l'albero cresce fino a quando tutte le foglie contengono meno di min_samples_split campioni).

### 5. `max_features`: None

- **Descrizione**: Il numero massimo di caratteristiche da considerare per la migliore suddivisione.
- **Funzione**: Limita il numero di caratteristiche utilizzate in ogni nodo per ridurre la complessità del modello.
- **Valore di Default**: None (tutte le caratteristiche sono considerate).

### 6. `max_leaf_nodes`: None

- **Descrizione**: Il numero massimo di foglie.
- **Funzione**: Limita il numero di nodi terminali (foglie) per prevenire l'overfitting.
- **Valore di Default**: None (illimitato).

### 7. `min_impurity_decrease`: 0.0

- **Descrizione**: Un nodo verrà suddiviso se questa divisione induce una diminuzione dell'impurità maggiore o uguale a questo valore.
- **Funzione**: Controlla quando fermare le suddivisioni per prevenire l'overfitting.
- **Valore di Default**: 0.0.

### 8. `min_samples_leaf`: 1

- **Descrizione**: Il numero minimo di campioni richiesti per essere in una foglia.
- **Funzione**: Controlla la dimensione minima di una foglia per evitare che l'albero cresca troppo in profondità.
- **Valore di Default**: 1.

### 9. `min_samples_split`: 2

- **Descrizione**: Il numero minimo di campioni richiesti per suddividere un nodo interno.
- **Funzione**: Controlla quando fermare le suddivisioni per evitare che l'albero cresca troppo in profondità.
- **Valore di Default**: 2.

### 10. `min_weight_fraction_leaf`: 0.0

- **Descrizione**: La frazione ponderata minima della somma totale dei pesi (tutti i campioni), richiesta per essere in una foglia.
- **Funzione**: Come `min_samples_leaf`, ma espressa come frazione dei pesi totali.
- **Valore di Default**: 0.0.

### 11. `random_state`: 42

- **Descrizione**: Il seme utilizzato dal generatore di numeri casuali.
- **Funzione**: Garantisce la riproducibilità dei risultati. Utilizzando lo stesso valore di `random_state`, si ottiene sempre la stessa sequenza di suddivisioni.
- **Valore di Default**: None (utilizza un generatore casuale diverso ad ogni esecuzione).

### 12. `splitter`: 'best'

- **Descrizione**: La strategia utilizzata per scegliere la suddivisione in ogni nodo.
- **Funzione**: Può essere 'best' per scegliere la migliore divisione o 'random' per scegliere una divisione casuale tra le migliori.
- **Valore di Default**: 'best'.
