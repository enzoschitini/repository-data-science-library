## NERD DOS DADOS - Trabalhando com arquivos Gigantes em Python

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('dados.csv', delimiter = ';', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Vamos avaliar quanto está sendo usado da memória do nosso computador para cada variável
# Importante: Usar o deep = True para que seja feito uma analise profunda e traga realmente as informações precisas
memoria = df.memory_usage(deep = True)    
```

```python
memoria
```

```python
memoria.sum()
```

```python
# Vamos achar a quantidade em Gb (GigaBytes) que está sendo utilizado da nossa RAM
# Temos a informação em Bytes, então precisamos converter até chegar em GigaBytes

# 1.024 Bytes = 1 Kb
# 1.024 Kb = 1 Mb
# 1.024 Mb = 1 Gb

# Então faremos a quantidade dos valores dividido por 1024 ao cubo
Memoria_Gb = memoria.sum() / (1024 ** 3)
Memoria_Gb
```

```python
# Avaliando o percentual de cada coluna
memoria / memoria.sum() * 100
```

```python
# Repare o seguinte
# Variavel DATA_CONTRATO é do tipo Date
# As outras variaveis são do tipo categóricas...

# Vamos converter a variavel DATA_CONTRATO para o tipo DateTime do Pandas
df["DATA_CONTRATO"] = pd.to_datetime(df["DATA_CONTRATO"])


# Variavel CLIENTE e FIADOR iremos descarta-la, pois sabemos que não será necessário, pois já temos um numero
# de contrato representando o cliente e seu fiador
df = df.drop("CLIENTE", axis=1)
df = df.drop("FIADOR", axis=1)


# Vamos converter as outras variaves que analisamos para o tipo categorico
df["UF_CLIENTE"] = df["UF_CLIENTE"].astype("category")
df["CIDADE_CLIENTE"] = df["CIDADE_CLIENTE"].astype("category")
df["SEXO"] = df["SEXO"].astype("category")
df["ESCOLARIDADE"] = df["ESCOLARIDADE"].astype("category")
df["CASA_PROPRIA"] = df["CASA_PROPRIA"].astype("category")
df["OUTRO_CONTRATO_ABERTO"] = df["OUTRO_CONTRATO_ABERTO"].astype("category")
df["OUTRO_CONTRATO_QUITADO"] = df["OUTRO_CONTRATO_QUITADO"].astype("category")
df["TRABALHA_ATUALMENTE"] = df["TRABALHA_ATUALMENTE"].astype("category")
```

```python
df.info()
```

```python
# Vamos avaliar novamente nossa memória
memoria = df.memory_usage(deep = True)  
Memoria_Gb2 = memoria.sum() / (1024 ** 3)
Memoria_Gb2
```
